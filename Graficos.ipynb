```python
import numpy as np

x = [10, 40, 50, 70]
x1 = np.array(x)
x2 = x1 * 2
y1 = [2, 4,6, 8]
y2 = np.array(y1) * 2
plt.figure(1) # Se crea la primera figura
plt.subplot(2,1,1) # vamos a crear dos gráficos en 2 filas, en una columna, primer grafico
plt.plot(x1,y1)
plt.subplot(2,1,2)  # vamos a crear dos gráficos en 2 filas, en una columna, segundo grafico
plt.plot(x2, y2, "g*")
plt.show()
```

```python
# Gráficos de Barras
```

```python
x = np.array(["Manzanas", "Peras", "Platanos", "Naranjas"])
y = np.array([4, 9, 3, 8])

plt.bar(x,y) # En vertical
plt.show()
```

```python
x = np.array(["Manzanas", "Peras", "Platanos", "Naranjas"])
y = np.array([4, 9, 3, 8])

plt.barh(x,y) # En horizontal cambiar y por x
plt.show()
```
