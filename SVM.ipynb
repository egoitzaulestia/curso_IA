# Vamos a crear dos grupos de datos

```python
from sklearn.datasets import make_blobs
```

```python
# Vamos a crear 2 nubes de puntos
X, y = make_blobs([25, 25], # Creamos 25 y 25
                 centers = [(1.5, 0), [-1.5, 1]], # Creamos los centros
                 cluster_std = [0.4, 0.4],
                 random_state = 5) # La semilla
```

# Vamos a ver la distribuci√≥n de los datos
Vamos a crear datos por nosotros mismos

```python
import fronteras
```

```python
fronteras.mostrar_fronteras(None, X, None, y, ["Clase 0", "Clase 1"]) # Esto es lo mismo que cuando predeciamos las areas, pero como no le hemos metido el modelo, se pasa a la segunda parte que es plotear los datos
```

```python
# SVC (Support Vector Classifier)
from sklearn.svm import SVC
```

```python
model = SVC(kernel = "linear")
model.fit(X, y) # Alimentamos y entrenamos el modelo
```

Ya tenemos nuestro modelo entrenado

Atributos m√°s importantes de  esta SVC:

    + .classes_
    + .support_
    + .support_vectors_ # tienen un x1, x2 e y1
    + .n_support_

```python
import pandas as pd
import seaborn as sb
iris = sb.load_dataset("iris")
```

```python
iris["label"] = iris.species.astype("category").cat.codes
```

```python
X = iris[["sepal_length", "sepal_width"]]
y = iris["label"]
fronteras.mostrar_fronteras(None, X.values, None, y, None, iris.species.unique()) # .unique() cogemos todas las clases de "label", porque coje todas las distintas posibles
```

```python
model = SVC(kernel = "linear")
model.fit(X.values, y) # Ene este ejercicio hemos hecho de forma r√†oida todos los datos juntos, pero en realidad SIEMPRE HAY QUE SEPARAR 80/20, ENTRENAMIENTO/TEST
model.score(X.values, y) # La funci√≥n .score() nos dice lo bueno que es la predicci√≥n del modelo
# En realidad caundo queremos hacer un score se lo hacemos con los vectores de TEST (en este ejercicio ni hemos separado)
# Osea
```
