```python
import seaborn as sb
iris = sb.load_dataset("iris")
iris["etiketa"] = iris.species.astype("category").cat.codes # vamos a crear una columna, lo hacemos categoricamente
# Emos añadido la columan etiketa
iris.head()
```

Fronteras de decisión para 2 clases: Seleccionamos las clases "versicolor" y "virginica" de las columna species. Dejamos "setosa" fuera.

```python
iris = iris[iris.species.isin(["versicolor", "virginica"])] # isin() "las que westan es"
# nuestro nuevo iris tendrá solo 2 columnas
```

```python
iris.head()
```

TEXTO...

```python
Hasta ahora hemos cogida dos filas, ahora vamos a coger 2 columnas
```

```python
X = iris[["sepal_length", "sepal_width"]] # "X" son siempre las variables predictoras. ogemos solo estas dos columanas, y las metemos en "X"
y = iris.etiketa # La "y" son siempre las variables en objetivo
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
from sklearn.linear_model import LogisticRegression
mrl = LogisticRegression()
mrl.fit(X_train.values, y_train.values)
```

```python
type(X_train.values) # nos devuleve un array multidimensional
```

Frontera de Decisión.

- Vamos a crear un módulo llamado "fronteras". Incluirá dos funciones:
    + plot_fronteras_decision, que muestra una gráfica con las fronteras de decisión de un modelo dde clasificación
    + mostrar_fronteras, que invoca a la función anterior y mouestra sobre ell un diagrama de dispersión con los datos de entrenamiento y de validación

```python
from fronteras import plot_fronteras_decision, mostrar_fronteras
```

```python
import numpy as np
```

```python
mostrar_fronteras(mrl, X_train.values, X_test.values, y_train, y_test, iris.species.unique())
```

```python
iris2 = sb.load_dataset("iris")
iris2 = iris2[iris2.species.isin(["setosa", "virginica"])]
iris2["etiketa"] = iris2.species.astype("category").cat.codes
iris2.head()
```
