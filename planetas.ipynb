```python
# Agrupaciones de años por "method" , para ver la primera vez que se usó cada uno
```

```python
planets.groupby("method")["year"].min()
```

```python
planets.groupby("method")["year"].describe()
```

```python
# VECTOR: Vamos a crear el vector decada
```

```python
decada = 10 * (planets["year"]//10)
# cociente de la división entera con "//"
decada
```

```python
decada = "dec " + decada.astype(str)
decada.name = "decada"
decada
```

```python
decada.describe()
```

```python
# Vamos a analizar el número de descubrimientos por método y década
```

```python
planets.groupby(["method", decada])["number"].sum()
```

```python
planets.groupby(["method", decada])["number"].sum().unstack()
```

```python
planets.groupby(["method", decada])["number"].sum().unstack().fillna('-') # Rellenamos los NAN con un 0
```
